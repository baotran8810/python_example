# Built-in Data Types
In programming, data type is an important concept.

Variables can store data of different types, and different types can do different things.

Python has the following data types built-in by default, in these categories:

- Text Type:	str 

- Numeric Types:	int, float, complex 

- Sequence Types:	list, tuple, range

- Mapping Type:	dict

- Set Types:	set, frozenset

- Boolean Type:	bool

- Binary Types:	bytes, bytearray, memoryview

- None Type:	NoneType

### Text type - String - str

```python
text = "Hello, Chang chang"
# f before the string is used to format the string
print(f"String (str): {text}- Type: {type(text)}")

```

Escape Sequence	Description



```python
print("Single Quote: \'Hello\'")      # Single quote
print("Double Quote: \"World\"")      # Double quote
print("Backslash: This is a backslash -> \\")  # Backslash
print("Newline:\nThis is on a new line")  # Newline
print("Tab:\tThis text is tabbed")    # Tab
print("Carriage Return: Hello\rWorld")  # Carriage Return
print("Backspace: ABC\bD")            # Backspace (removes 'C')# Hex values (A B C)

# Using a raw string to ignore escape sequences
print(r"Raw String: C:\Users\Name\Documents")
```

1. String Case Conversion

```python
text = "Hello, Chang chang"

print(text.upper())
print(text.lower())
print(text.title())
print(text.capitalize())
print(text.swapcase())
```

2. String Trimming (Remove Whitespace)

```python
text = "  hello world  "
print(text.strip())   # "hello world"
print(text.lstrip())  # "hello world  "
print(text.rstrip())  # "  hello world"

print(text.replace(" ", ""))
```

3. Checking String Content

```python
text = "Hello123"

print(text.isalpha())   # False (contains numbers)
print(text.isdigit())   # False (contains letters)
print(text.isalnum())   # True (letters and numbers only)
print(text.isspace())   # False (not just spaces)
print(text.islower())   # False (has uppercase)
print(text.isupper())   # False (has lowercase)
print(text.startswith("Hello"))  # True
print(text.endswith("123"))  # True
```

4. String Searching

```python
text = "Hello, welcome to Python!"

print(text.find("Python"))  # 18 (index where it starts)
print(text.index("Python"))  # 18 (same as find, but raises error if not found)
print(text.count("o")) 
```

5. String Replacement

```python
text = "I love Java"
print(text.replace("Java", "Python"))  # "I love Python"
```

6. Splitting and Joining Strings

```python
text = "apple,banana,grape"
words = text.split(",")  # ['apple', 'banana', 'grape']
print(words)

new_text = "-".join(words)  # "apple-banana-grape"
print(new_text)
```

7. String Formatting (f-strings)


```python
name = "Alice"
age = 25
print(f"My name is {name} and I am {age} years old.")

```

8. String Concatenation


```python
a = "Hello"
b = "World"
c = a + " " + b
print(c)
```
